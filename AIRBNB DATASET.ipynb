```python
airbnb.drop(["latitude","longitude","last_review","reviews_per_month"],axis=1,inplace=True)
```

```python
host_area=airbnb.groupby(["host_name","neighbourhood_group"])["calculated_host_listings_count"].max().reset_index()
```

```python
host_area.sort_values(by="calculated_host_listings_count",ascending=False).head()
```

```python
sns.pairplot(airbnb)
plt.show()
```

```python
room_price_area_wise=airbnb.groupby(["neighbourhood_group","room_type"])["price"].max().reset_index()
```

```python
room_price_area_wise.sort_values(by="price",ascending=False).head(10)
```

```python
plt.bar(list(airbnb["room_type"].value_counts().keys()),list(airbnb["room_type"].value_counts()),color=["g","r","b"])
plt.show()
```

```python
plt.bar(list(airbnb["neighbourhood_group"].value_counts().keys()),list(airbnb["neighbourhood_group"].value_counts()),color=["g","r","b","orange","pink"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(airbnb["neighbourhood"].value_counts()[0:5].keys()),list(airbnb["neighbourhood"].value_counts()[0:5]),color=["g","r","b","orange","pink"])
plt.show()
```

```python
area_review=airbnb.groupby(["neighbourhood_group"])["number_of_reviews"].max().reset_index()
area_review
```

```python
area=area_review["neighbourhood_group"]
review=area_review["number_of_reviews"]
```

```python
plt.bar(area,review,color=["g","r","b","orange","pink"],width=0.5)
plt.xlabel("area")
plt.ylabel("review")
plt.title("Number of review in terms of area")
plt.show()
```

```python
area_price=airbnb.groupby(["price"])["number_of_reviews"].max().reset_index()
area_price.head(10)
```

```python
price_list=area_price["price"]
price_review=area_price["number_of_reviews"]
```

```python
plt.scatter(price_list,price_review)
plt.xlabel("price")
plt.ylabel("number_of_reviews")
plt.title("Number of Reviews Vs Price")
plt.show()
```

```python
busy_host=airbnb.groupby(["host_id","host_name","room_type"])["number_of_reviews"].max().reset_index()
busy_host.sort_values(by="number_of_reviews",ascending=False).head(10)
```

```python
name_host=busy_host["host_name"]
review_host=busy_host["number_of_reviews"]
```

```python
plt.bar(name_host[0:10],review_host[0:10],color=["g","r","b","orange","pink"])
plt.xlabel("Name of the Host")
plt.ylabel("Review")
plt.title("Busiest Host in terms of reviews")
plt.show()
```

```python
highest_price=airbnb.groupby(["host_id","host_name","room_type","neighbourhood_group"])["price"].max().reset_index()
highest_price.sort_values(by="price",ascending=False).head(10)
```

```python
host_price=highest_price["host_name"]
host_review=highest_price["price"]
```

```python
plt.figure(figsize=(6,4))
plt.bar(host_price[0:10],host_review[0:10],color=["g","r","b","orange","pink"])
plt.xlabel("Name of the Host")
plt.ylabel("Review")
plt.title("Busiest Host in terms of reviews")
plt.show()
```
